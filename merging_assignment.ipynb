```python
import pandas as pd
import csv
```

```python
m1 = pd.read_csv('example1.csv')
m2 = pd.read_csv('example2.csv')
```

```python
def deleting_unmatched(data1,data2):
    matched_chr = data1[data1['cor1'].isin(data2['cor1']) & data1['cor2'].isin(data2['cor2'])]
    return matched_chr

new_data = deleting_unmatched(m1,m2)
new_data
```

```python
new_data.to_csv('solved.csv', index=False)
```
