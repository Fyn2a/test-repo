```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
s1 = "AGCTTTTCATTCTGACTGCAACGGGCAATATGTCTCTGTGTGGATTAAAAAAAGAGTGTCTGATAGCAGC"
countA = 0
countC = 0
countG = 0
countT = 0
for i in s1:
    print(" ")
    if i == "A":
        countA = countA+1
    elif i == "C":
        countC = countC+1
    elif i == "G":
        countG = countG+1
    elif i == "T":
        countT = countT+1
        

print("Count of A:", countA)
print("Count of C:", countC)
print("Count of G:", countG)
print("Count of T:", countT)
    
        
```
