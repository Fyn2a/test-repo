**Q8: Count the number of countries for which the `total_tests` data is missing.**

*Hint: Use the `.isna` method.*

```python
covid_data_df.head()
```

```python
#we can solve it in one line by using the method isna() & sum()
print("The data for total tests is missing for {} countries.".format(covid_data_df['total_tests'].isna().sum()))

#or by using for loop
total_tests_missing2 = 0
for i in covid_data_df['total_tests']:
    if pd.isna(i):
        total_tests_missing2 += 1
    else:
        continue
    
print("The data for total tests is missing for {} countries.".format(int(total_tests_missing2)))
```

Let's merge the two data frames, and compute some more metrics.

**Q9: Merge `countries_df` with `covid_data_df` on the `location` column.**

*Hint: Use the `.merge` method on `countries_df`.

```python
combined_df = countries.merge(covid_data_df, how = 'right', on='location')
combined_df
```

**Q10: Add columns `tests_per_million`, `cases_per_million` and `deaths_per_million` into `combined_df`.**

```python
combined_df['tests_per_million'] = (combined_df['total_tests']/combined_df['population']) * 1e6 #I think "per million" here refers to per million people that's why I dividied by population
```

```python
combined_df['cases_per_million'] = (combined_df['total_cases']/combined_df['population']) * 1e6
```

```python
combined_df['deaths_per_million'] = (combined_df['total_deaths']/combined_df['population']) * 1e6
```

```python
combined_df
```

**Q11: Create a dataframe with 10 countires that have highest number of tests per million people.**

```python
list_TestM = combined_df.sort_values(by='tests_per_million', ascending=False).head(10) 
#display(list_TestM)

countries_sorted2 = [locSor for locSor in list_TestM['location']] #this list comprehension is to extract only the coutries names

highest_tests_df = pd.DataFrame(countries_sorted2,
                                columns=['countries']
                                )   
display(highest_tests_df)
```

**Q12: Create a dataframe with 10 countires that have highest number of positive cases per million people.**

```python
listCm = combined_df.sort_values(by='cases_per_million', ascending=False).head(10)

countries_sorted3 = [locSor for locSor in listCm['location']]

highest_cases_df = pd.DataFrame(countries_sorted3,
                                columns=['countries']
                                )  
display(highest_cases_df)
```

**Q13: Create a dataframe with 10 countires that have highest number of deaths cases per million people?**

```python
highest_deaths_df = combined_df.sort_values(by='deaths_per_million', ascending=False).head(10)
display(highest_deaths_df)
```
