```python
import csv
import re

def txt_to_csv(txt_input, csv_output):
    with open(txt_input, 'r') as infile, open(csv_output, 'w', newline='') as outfile:
        writer = csv.writer(outfile)
        
        for line in infile:
            if line.startswith('#CHROM'):  
                headers = re.split(r'\s+', line.strip()) #split header by any whitespace
                writer.writerow(headers)  
                break
        
        for line in infile:
            row = re.split(r'\s+', line.strip())
            writer.writerow(row)  

txt_to_csv("vcf_1.txt", "vcf_1.csv")
txt_to_csv("vcf_2.txt", "vcf_2.csv")


```

```python
m1 = pd.read_csv('vcf_1.csv')
m2 = pd.read_csv('vcf_2.csv')
```
